```python
import joblib
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
import copy
```

```python
diamond_price_prediction=joblib.load("Diamond_Prediction_Model")
```

```python
pred_data_df=pd.read_csv("test.csv")
```

```python
pred_data_df
```

```python
pred_data_copy=copy.deepcopy(pred_data_df)
```

```python
pred_data_copy
```

```python
pred_data_copy=pred_data_copy.drop("id",axis=1)
```

```python
pred_data_copy
```

```python
l_enc=LabelEncoder()
```

```python
l_cut=l_enc.fit_transform(pred_data_copy.cut)
```

```python
l_color=l_enc.fit_transform(pred_data_copy.color)
```

```python
l_clarity=l_enc.fit_transform(pred_data_copy.clarity)
```

```python
pred_data_copy["l_cut"]=l_cut
```

```python
pred_data_copy["l_color"]=l_color
```

```python
pred_data_copy["l_clarity"]=l_clarity
```

```python
pred_data_copy
```

```python
pred_data_copy=pred_data_copy.drop(["cut","color","clarity"],axis=1)
```

```python
pred_data_copy
```

```python
sca_data=StandardScaler()
```

```python
sca_data=sca_data.fit_transform(pred_data_copy)
```

```python
sca_data
```

```python
diamond_price_prediction.predict(sca_data)
```

```python
pred_data_df["price"]=diamond_price_prediction.predict(sca_data)
```

```python
pred_data_df.head(2)
```

```python
pred_data_df.to_csv("predicted_price.csv",index=False,columns=["id","price"])
```
